```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Dataset.csv")

print(df.head())
print(df.shape)
```

```python
df = df[['City','Locality','Longitude','Latitude',
         'Aggregate rating','Cuisines','Price range']]
```

```python
print(df.isnull().sum())
```

```python
df = df.dropna()
```

```python
print(df.isnull().sum())
```

```python
city_counts = df['City'].value_counts()

plt.figure()
city_counts.head(10).plot(kind='bar')
plt.title("Top 10 Cities by Restaurant Count")
plt.xticks(rotation=45)
plt.show()
```

```python
avg_rating_city = df.groupby('City')['Aggregate rating'].mean()

plt.figure()
avg_rating_city.sort_values(ascending=False).head(10).plot(kind='bar')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure()
sns.countplot(x='Price range', data=df)
plt.show()
```

```python
top_cuisines = df['Cuisines'].value_counts().head(10)

plt.figure()
top_cuisines.plot(kind='bar')
plt.xticks(rotation=45)
plt.show()
```

```python
rating_price = df.groupby('Price range')['Aggregate rating'].mean()

plt.figure()
rating_price.plot(kind='bar')
plt.show()
```

```python
plt.figure()
plt.scatter(df['Longitude'], df['Latitude'])
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("Restaurant Location Distribution")
plt.show()
```

```python
print("Key Insights:")
print("1. Certain cities have higher restaurant concentration.")
print("2. Mid-range restaurants tend to have better average ratings.")
print("3. Popular cuisines dominate metro cities.")
print("4. Restaurants are clustered in high population areas.")
```
